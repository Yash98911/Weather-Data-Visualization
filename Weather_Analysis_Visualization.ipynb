```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = np.arange(1,31)
max_temp = np.array([47, 36, 41, 43, 39, 39, 39, 45, 35, 41, 43, 40, 41, 40, 38, 36, 47, 42, 48, 38, 48, 40, 40, 35,
 47, 42, 48, 43, 48, 45])
min_temp = np.array([19, 17, 17, 18, 16, 19, 17, 15, 15, 16, 15, 17, 19, 15, 18, 16, 16, 16, 15, 17, 18, 15, 18, 19,
 17, 17, 19, 15, 18, 17])
avg_temp = (max_temp+min_temp)//2
rainfall = np.array([23, 23,  5, 22, 10,  2,  6, 20, 20, 14,  9, 22, 12, 18, 10, 16,  3, 14,  5, 12,  3,  0, 15,  0,
  3,  4, 18,  2,  1, 15])
humidity = np.array([79, 87, 37, 53, 45, 75, 28, 25, 87, 89, 65, 36, 62, 64, 55, 41, 57, 31, 86, 67, 69, 53, 28, 85,
 39, 22, 52, 62, 67, 38])
```

```python
plt.figure(figsize=(12,6))
plt.plot(days,max_temp,marker="o",color="r",label = "Max Temp")
plt.plot(days,avg_temp,marker="*",color="g",label="Avg Temp")
plt.plot(days,min_temp,marker="<",color="b",label="Min Temp")
plt.legend(loc=2)
plt.xticks(days)
plt.xlabel("Days",color="purple",fontsize=15)
plt.ylabel("Temprature",color="purple",fontsize=15)
plt.title("Variation of temprature in a month",fontsize=15,color="purple")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(days,rainfall,color="r",label = "Amount of rainfall",alpha = 0.7,edgecolor="black")
plt.xticks(days)
plt.xlabel("Days",color="purple",fontsize=15)
plt.ylabel("Rainfall(mm)",color="purple",fontsize=15)
plt.title("Rainfall in different days of month",color="purple",fontsize=15)
for i in range(len(days)):
    plt.text(days[i],rainfall[i]+1,str(rainfall[i]),color="black",ha="center")
plt.ylim(0,max(rainfall)+10)
plt.legend()
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(12,6))
plt.plot(days,humidity,color="r",marker = "o",label="Humidity")
plt.xticks(days)
plt.xlabel("Days",color="purple",fontsize=15)
plt.ylabel("Humidity",color="purple",fontsize=15)
plt.title("Humidity change over the month",fontsize=15,color="purple")
plt.legend()
plt.grid()
max_index = np.argmax(humidity)
min_index = np.argmin(humidity)
plt.text(days[max_index], humidity[max_index] + 2, f"Max: {humidity[max_index]}%", color="black", ha="center")
plt.text(days[min_index],humidity[min_index] - 4,f"Min: {humidity[min_index]}%",color="black",ha="center")
plt.ylim(0,95)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.hist(avg_temp,color="r",label="Avg Temp",bins=6,edgecolor="black",alpha=0.8)
plt.xlabel("Temprature",color="purple",fontsize=15)
plt.ylabel("Frequency",color="purple",fontsize = 15)
plt.title("Frequency of temprature over the month",color="purple",fontsize=15)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

days = np.arange(1,31)
max_temp = np.array([47, 36, 41, 43, 39, 39, 39, 45, 35, 41, 43, 40, 41, 40, 38, 36, 47, 42, 48, 38, 48, 40, 40, 35,
 47, 42, 48, 43, 48, 45])
min_temp = np.array([19, 17, 17, 18, 16, 19, 17, 15, 15, 16, 15, 17, 19, 15, 18, 16, 16, 16, 15, 17, 18, 15, 18, 19,
 17, 17, 19, 15, 18, 17])
avg_temp = (max_temp+min_temp)//2
rainfall = np.array([23, 23,  5, 22, 10,  2,  6, 20, 20, 14,  9, 22, 12, 18, 10, 16,  3, 14,  5, 12,  3,  0, 15,  0,
  3,  4, 18,  2,  1, 15])
humidity = np.array([79, 87, 37, 53, 45, 75, 28, 25, 87, 89, 65, 36, 62, 64, 55, 41, 57, 31, 86, 67, 69, 53, 28, 85,
 39, 22, 52, 62, 67, 38])

max_index = np.argmax(max_temp)
min_index = np.argmin(min_temp)

plt.figure(figsize=(12,6))
plt.plot(days,max_temp,marker="o",color="r",label = "Max Temp")
plt.plot(days,avg_temp,marker="*",color="g",label="Avg Temp")
plt.plot(days,min_temp,marker="<",color="b",label="Min Temp")
plt.legend(loc=2)
plt.xticks(days)
plt.xlabel("Days",color="purple",fontsize=15)
plt.ylabel("Temprature",color="purple",fontsize=15)
plt.title("Variation of temprature in a month",fontsize=15,color="purple")
plt.text(days[max_index],max_temp[max_index] + 2,f"Max temp: {max_temp[max_index]}C",ha="center",fontsize = 10)
plt.text(days[min_index],min_temp[min_index]-4,f"Min temp: {min_temp[min_index]}C",ha="center",fontsize=10)
plt.ylim(0,60)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig , axs = plt.subplots(2,2,figsize = (14,10))
axs[0,0].plot(days,max_temp,marker="o",color="r",label = "Max Temp")
axs[0,0].plot(days,avg_temp,marker="*",color="g",label="Avg Temp")
axs[0,0].plot(days,min_temp,marker="<",color="b",label="Min Temp")
axs[0,0].legend()
axs[0,0].set_title("Temprature Trend")


weather_type = ["Sunny","Rainy","Humid"]
count = [12,8,10]
axs[0,1].pie(count,labels=weather_type,colors=["r","g","yellow"],autopct = "%1.1f%%",shadow=True,explode = [0.05,0.05,0.05],startangle = 90)
axs[0,1].legend()
axs[0,1].set_title("Types of weather across the month",color="magenta",fontsize=15)


axs[1,0].bar(days,rainfall,color="r",label = "Amount of rainfall")
axs[1,0].set_title("Rainfall in different days of month",color="purple",fontsize=15)
axs[1,0].legend()


hottest_day = days[np.argmax(max_temp)]
coldest_day = days[np.argmin(min_temp)]
max_rain_day = days[np.argmax(rainfall)]
total_rainfall = sum(rainfall)
avg_humidity = np.mean(humidity)

summary = f"""
Hottest Day: Day {hottest_day} ({max(max_temp)}¬∞C)
Coldest Day: Day {coldest_day} ({min(min_temp)}¬∞C)
Day with Max Rainfall: Day {max_rain_day} ({max(rainfall)} mm)
Total Monthly Rainfall: {total_rainfall} mm
Average Humidity: {avg_humidity:.1f}%
"""

axs[1, 1].axis("off")  
axs[1, 1].text(0, 0.5, summary, fontsize=12, color="black", va="center")
axs[1, 1].set_title("Summary Report", fontsize=15, color="purple")
plt.show()
```
